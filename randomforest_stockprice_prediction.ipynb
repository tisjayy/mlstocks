```python
import yfinance as yf
import pandas as pd
import pytz
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from sklearn.model_selection import RandomizedSearchCV # Import RandomizedSearchCV

def fetch_and_prepare_data(ticker, start_date=None):
    """Fetches historical data for a given ticker and prepares it for the model.
       Note: Using period="5y" might affect features requiring longer history."""
    stock_data = yf.Ticker(ticker)
    # Fetching data for 10 years
    stock_data = stock_data.history(period="10y", interval ="1d")

    # Drop unnecessary columns
    if 'Dividends' in stock_data.columns:
        del stock_data['Dividends']
    if 'Stock Splits' in stock_data.columns:
        del stock_data['Stock Splits']
    if 'Capital Gains' in stock_data.columns:
        del stock_data['Capital Gains']

    # Add 'Tomorrow' and 'Target' columns
    stock_data["Tomorrow"] = stock_data["Close"].shift(-1)
    stock_data["Target"] = (stock_data["Tomorrow"] > stock_data["Close"]).astype(int)

    # Add enhanced predictors
    horizons = {2, 5, 60, 250, 500} # Using horizons up to 500
    for horizon in horizons:
        rolling_averages = stock_data.rolling(horizon).mean()
        ratio_column = f"Close_Ratio_{horizon}"
        stock_data[ratio_column] = stock_data["Close"] / rolling_averages["Close"]
        trend_column = f"Trend_{horizon}"
        stock_data[trend_column] = stock_data.shift(1).rolling(horizon).sum()["Target"]

    # Add previously removed predictors back
    stock_data["Percent_Off_High"] = (stock_data["High"] - stock_data["Close"]) / stock_data["High"]
    stock_data["ATR_14"] = stock_data[["High", "Low", "Close"]].apply(lambda x: max(x["High"] - x["Low"], abs(x["High"] - x["Close"]), abs(x["Low"] - x["Close"])), axis=1).rolling(window=14).mean()
    stock_data["Volume_Surge"] = stock_data["Volume"].rolling(window=10).mean()
    stock_data["Return_1M"] = stock_data["Close"].pct_change(periods=20)
    stock_data["RS_1W_vs_20D"] = stock_data["Close"].rolling(window=5).mean() / stock_data["Close"].rolling(window=20).mean()

    # Add EMA features
    stock_data["EMA_midterm"] = stock_data["Close"].ewm(span=50, adjust=False).mean()
    stock_data["EMA_longterm"] = stock_data["Close"].ewm(span=200, adjust=False).mean()

    # Add MACD (using typical periods 12, 26, 9)
    stock_data["EMA_12"] = stock_data["Close"].ewm(span=12, adjust=False).mean()
    stock_data["EMA_26"] = stock_data["Close"].ewm(span=26, adjust=False).mean()
    stock_data["MACD"] = stock_data["EMA_12"] - stock_data["EMA_26"]
    stock_data["Signal_Line"] = stock_data["MACD"].ewm(span=9, adjust=False).mean()

    # Calculate RSI (Relative Strength Index) with period 14
    delta = stock_data["Close"].diff()
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    avg_gain = gain.ewm(span=14, adjust=False).mean()
    avg_loss = loss.ewm(span=14, adjust=False).mean()
    rs = avg_gain / avg_loss
    stock_data["RSI"] = 100 - (100 / (1 + rs))
    stock_data["RSI"].replace([float('inf'), float('-inf')], 100, inplace=True)
    stock_data["RSI"].fillna(0, inplace=True) # Handle potential NaN values in RSI

    # Add new binary RSI feature (35 <= RSI < 70)
    stock_data["RSI_above_70"] = (stock_data["RSI"] > 70).astype(int)    # Overbought
    stock_data["RSI_below_30"] = (stock_data["RSI"] < 30).astype(int)    # Oversold
    stock_data["RSI_mid_range"] = ((stock_data["RSI"] >= 30) & (stock_data["RSI"] <= 70)).astype(int)
    stock_data["RSI_above_60"] = (stock_data["RSI"] > 50).astype(int)

    # Add Bollinger Bands (using typical period 20 and 2 standard deviations)
    stock_data["Rolling_Mean_20"] = stock_data["Close"].rolling(window=20).mean()
    stock_data["Rolling_Std_20"] = stock_data["Close"].rolling(window=20).std()
    stock_data["Upper_Band"] = stock_data["Rolling_Mean_20"] + (stock_data["Rolling_Std_20"] * 2)
    stock_data["Lower_Band"] = stock_data["Rolling_Mean_20"] - (stock_data["Rolling_Std_20"] * 2)

    # Step 1: Identify RSI in mid range (43–57)
    mid_range = stock_data["RSI"].between(43, 57)

    # Step 2: Detect new entries into the mid range (from outside)
    entered_mid_range = (mid_range & ~mid_range.shift(1, fill_value=False)).astype(int)

    # Step 3: Get index positions of entry points
    entry_indices = entered_mid_range[entered_mid_range == 1].index.tolist()

    # Step 4: Initialize the signal column
    stock_data["RSI_Third_Entry_Bull"] = 0

    # Proceed only if at least 3 entries occurred
    if len(entry_indices) >= 3:
        first_entry_idx = entry_indices[0]
        third_entry_idx = entry_indices[2]

        # Check if RSI was between 30–35 before the first entry (look back 10 days)
        rsi_before_first = stock_data.loc[:first_entry_idx].iloc[-11:-1]["RSI"]  # last 10 days before first entry
        came_from_30_35 = ((rsi_before_first >= 30) & (rsi_before_first <= 35)).any()

        if came_from_30_35:
            # Set signal at third entry index
            stock_data.loc[third_entry_idx, "RSI_Third_Entry_Bull"] = 1


    # Slice the data to the last 10 years *before* dropping NaNs
    if start_date:
        # Localize the start_date to the timezone of the DataFrame index
        if stock_data.index.tz is not None:
            local_tz = stock_data.index.tz
            start_date = start_date.replace(tzinfo=pytz.utc).astimezone(local_tz)
        stock_data = stock_data.loc[start_date:]

    # Drop rows with NaN values created by rolling calculations
    stock_data.dropna(inplace=True)

    return stock_data

def predict(train, test, predictors, model):
    """Trains the model and returns probability predictions."""
    model.fit(train[predictors], train["Target"])
    # Return the probability of the positive class (stock going up)
    preds_proba = model.predict_proba(test[predictors])[:, 1]
    preds_proba = pd.Series(preds_proba, index=test.index, name="Probability_Up")
    return preds_proba

def backtest(data, model, predictors, start=300, step=250):
    """Performs backtesting of the model."""
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        model.fit(train[predictors], train["Target"])
        # Get the probability predictions - Pass the model argument
        predictions_proba = predict(train, test, predictors, model)
        # Concatenate the target and probability predictions
        combined = pd.concat([test["Target"], predictions_proba], axis=1)
        all_predictions.append(combined)

    return pd.concat(all_predictions)

def tune_hyperparameters(data, predictors, n_iter=20):
    """Tunes RandomForest hyperparameters using RandomizedSearchCV."""
    rf = RandomForestClassifier(random_state=42, n_jobs=-1)
    param_dist = {
        "n_estimators": [100, 200, 300, 500],
        "max_depth": [None, 10, 20, 30, 40, 50],
        "min_samples_split": [2, 5, 10],
        "min_samples_leaf": [1, 2, 4],
        "max_features": ["sqrt", "log2", None],
        "bootstrap": [True, False]
    }
    search = RandomizedSearchCV(
        rf, param_distributions=param_dist,
        n_iter=n_iter, scoring="precision",
        cv=3, random_state=42, n_jobs=-1
    )
    # Use the first 70% of data for tuning to avoid leakage
    train_split = int(0.7 * len(data))
    search.fit(data[predictors][:train_split], data["Target"][:train_split])
    return search.best_params_

def train_and_predict(data, predictors, start=2500, step=250):
    """Tunes the RF model, then trains and backtests with the best parameters."""
    print("[INFO] Starting hyperparameter tuning...")
    best_params = tune_hyperparameters(data, predictors)
    print("[INFO] Best Parameters Found:", best_params)

    model = RandomForestClassifier(**best_params, random_state=42)
    predictions = backtest(data, model, predictors, start=start, step=step)
    predictions["Predictions"] = (predictions["Probability_Up"] >= 0.60).astype(int)
    return predictions, best_params

def evaluate_predictions(predictions):
    return precision_score(predictions["Target"], predictions["Predictions"])

def get_latest_prediction(data, predictors, best_params):
    """Trains the model with tuned parameters on full data and predicts next day."""
    model = RandomForestClassifier(**best_params, random_state=42)
    train_data = data.iloc[:-1]
    model.fit(train_data[predictors], train_data["Target"])
    latest_data_point = data.iloc[-1][predictors].values.reshape(1, -1)
    return model.predict_proba(latest_data_point)[:, 1][0]
```

```python
import yfinance as yf
import pandas as pd

# Fetch AAPL data using yfinance
aapl = yf.Ticker("AAPL")
df = aapl.history(period="10y")

# Reset the index and get the 'close' column
df2 = df.reset_index()['Close']

# Display the head of the dataframe
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
df2 = scaler.fit_transform(np.array(df2).reshape(-1,1))
df2
```

divide data based on date, as it is time series problem

```python
training_size = int(len(df2)*0.65)
test_size = len(df2)-training_size
train_data, test_data = df2[0:training_size,:],df2[training_size:len(df2),:1]
```

```python
training_size,test_size
```

```python
import numpy as np
def create_dataset(dataset, time_step=1):
    dataX, dataY = [], []
    for i in range(len(dataset)-time_step-1):
        a = dataset[i:(i+time_step), 0]
        dataX.append(a)
        dataY.append(dataset[i + time_step, 0])
    return np.array(dataX), np.array(dataY)
```

```python
time_step = 100;
X_train, y_train = create_dataset(train_data, time_step)
X_test, ytest = create_dataset(test_data, time_step)
```

```python
X_train = X_train.reshape(X_train.shape[0],X_train.shape[1] , 1)
X_test = X_test.reshape(X_test.shape[0],X_test.shape[1] , 1)
#reshaping to [samples, time steps, features] format for LSTM
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
model = Sequential()
model.add(LSTM(50,return_sequences=True,input_shape=(100,1)))
model.add(LSTM(50,return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss='mean_squared_error',optimizer='adam')
```

In the selected cell, Mean Squared Error (MSE) is used as the loss function because this LSTM model is set up for a regression task. We are trying to predict a continuous value, which is the future stock price. MSE measures the average squared difference between the actual and predicted values, which is suitable for this type of problem.

```python
model.fit(X_train,y_train,validation_data=(X_test,ytest),epochs=100,batch_size=64,verbose=1)
```

```python
import tensorflow as tf
train_predict=model.predict(X_train)
test_predict=model.predict(X_test)
```
